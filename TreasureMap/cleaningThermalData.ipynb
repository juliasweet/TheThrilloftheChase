```python
#Import dependencies

import pandas as pd
```

```python
# Import csv

dataFile = "thermalRawData.csv"
allThermalData = pd.read_csv(dataFile)
```

```python
allThermalDataDF = pd.DataFrame(allThermalData)
```

```python
allThermalDataDF.head()
```

```python
#Confirm state abbs
allThermalDataDF.Column1.unique()
```

```python
# Get rid of extra spaces.
allThermalDataDF['Column1'] = allThermalDataDF['Column1'].str.strip()
```

```python
allThermalDataDF.head()
```

```python
# Get rid of unneeded columns in new DF
allStatesDF = allThermalDataDF[['Column1', 'Column2.1', 'Column2.4', 'Column3.1', 'Column3.2', 'Column3.3', 'Column3.4']].copy()
```

```python
allStatesDF.head()
```

```python
# Rename columns
allStatesDF.rename(columns = {"Column1":"State", 
                                            "Column2.1":"Latitude",
                                           "Column2.4":"Longitude",
                                           "Column3.1" : "Name1",
                                           "Column3.2": "Name2",
                                           "Column3.3": "Name3",
                                           "Column3.4": "Name4"},
                                inplace = True)
```

```python
allStatesDF.head()
```

```python
# Create column with consolidated site names
allStatesDF["SpringName"] = allStatesDF["Name1"].str.cat(allStatesDF[["Name2", "Name3", "Name4"]], na_rep = ' ', sep = ' ')
```

```python
allStatesDF.head()
```

```python
allStatesConcise = allStatesDF[["State", "Latitude", "Longitude", "SpringName"]]
```

```python
allStatesConcise.head()
```

```python
# Create array of states for filter
rockiesStates = ['NM', 'CO', 'WY', "MT"]
```

```python
rockies1 = allStatesConcise[allStatesConcise.State.isin(rockiesStates)]
```

```python
rockies1.head()
```

```python
# Export to csv
rockies1.to_csv("rockiesHotSprings.csv", index=False)
```

```python
# Get rid of springs that are south of Santa Fe
rockies2 = rockies1.loc[rockies1["Latitude"] > 35.41, :]
```

```python
rockies2.head()
```

```python
# loc with contains?
warmSprings = rockies2.loc[rockies2['SpringName'].str.contains(r'WARM')==True]
```

```python
warmSprings.head()
```

```python
warmSprings.to_csv("WarmSprings.csv")
```
